```python
import preprocessor as p
import re

reg = '([@#][A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)'

def clean_tweet(tweet):
    p.set_options(p.OPT.URL, p.OPT.MENTION)
    return p.tokenize(tweet.lower()).replace('$MENTION$', '').replace('#', '').replace("'", '').replace('\\n', '').replace('\xf0', ' \xf0')

```
